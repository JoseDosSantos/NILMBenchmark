Main info about AMPds: Monitoring length of two years (730 days of captured data per meter). AMPds2 data has been cleaned to provide for consistent and comparable accuracy results amongst different researchers and machine learning algorithms. Each row in each of the dataset files represents a single meter reading once every minute with an associated unix timestamp.

Additionally the data set provides information about water and gas consumption, which can also be considered when we focus the overall purposes of saving resources and compare costs by analysing bill data included as well.

Testing the accuracy performance with real-world datasets is crucial in these fields of research. Synthesized data does not realistically represent an actual dataset as ‘a real-world dataset would normally have certain complexity that is harder predict and in many cases can be very
difficult to deal with'.

Description of appliances (only available online [here](https://www.nature.com/articles/sdata201637/tables/5)):

UNIX_TS: The unix timestamp integer value of a given reading. The house used in AMPds is in the America/Vancouver timezone. So the timestamp 1333263600 is equivalent to April 01, 2012 12:00:00 AM (local time) or Apr 01, 2012 07:00:00 (GMT or UTC). Timestamps having a plus sign in front of it (e.g., +1333263600) denotes that that reading had missing data and the record was created algorithmically.

```python
# check appliance usage (for the total timeframe)
app = amp.drop(['WHE', 'RSE', 'GRE', 'MHE', 'UNE', 'UNIX_TS'], axis = 1) # only contains individual appliances
app.sum().plot(kind='pie', autopct='%1.1f%%', figsize=(10,10))
```

The pie chart displays, that most of the power throughout the two years has been used for the heat pump (25.4%) and the air furnance fan (17.5%). Both named appliances might be highly correlated with the weather situation, such as weather conditions or temperature. 

```python
app.plot.line(subplots=True, layout=(9,2), figsize=(10,10))
```

This graphic shows, that some appliances were turned on for several weeks or even months, i.e. utility room plug (UTE), while others like the fridge (FGE) are turned on/off usually. Besides, the outside plug (OUE) only appears to be used during a specific timeframe, which again underlines how much other conditions such as weather or temperature influence the usage of individual appliances.

Here we can already see, that the appliances follow some regular behavior, which can be described as their individual signature. Mainly, devices can be divided into 4 categories: on/off (also called two-state appliances), continously-varying appliances, multi-state appliances and permanent appliances.

## Clustering

It involves automatically discovering natural grouping in data. Unlike supervised learning (like predictive modeling), clustering algorithms only interpret the input data and find natural groups or clusters in feature space.

When clustering take care of the time series: trendsm cycles and seasonal changes.

```python
# merge both dataframes on datetime
df = pd.merge(amp, weather, on='Date/Time', how='left')

# drop columns not used
df = df.drop(['Year', 'Month', 'Day', 'Time', 'Data Quality', 'Temp Flag', 'Dew Point Temp Flag', 
              'Rel Hum Flag', 'Wind Dir (10s deg)', 'Wind Dir Flag', 'Wind Spd Flag', 'Wind Spd (km/h)', 
              'Visibility (km)', 'Visibility Flag', 'Stn Press (kPa)', 'Stn Press Flag', 'Hmdx', 
              'Hmdx Flag', 'Wind Chill', 'Wind Chill Flag'], axis=1)

# fill in nas with latest non missing value (hourly)
df = df.ffill(axis = 0) 
```

```python
# create dataframe for clustering
cols_keep = ['WHE', 'Date/Time', 'Temp (C)', 'Rel Hum (%)', 'Weather']
total = df[cols_keep]
```

```python
df.Weather.value_counts() # encode objetc variable, i.e. rain = 0/1
```

```python
# rain
df['Rain'] = df['Weather'].map(lambda x: 1 if 'Rain' in str(x) else 0)

# snow
df['Snow'] = df['Weather'].map(lambda x: 1 if 'Snow' in str(x) else 0)

# clear or cloudy
df['Clear'] = df['Weather'].map(lambda x: 1 if 'Clear' in str(x) else 0)
df['Cloudy'] = df['Weather'].map(lambda x: 1 if 'Cloudy' in str(x) else 0)
```
